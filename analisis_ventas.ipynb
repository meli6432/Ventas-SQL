```python
query = """
SELECT c.nombre, p.nombre_producto, v.fecha_venta, v.total
FROM ventas v
JOIN clientes c ON v.cliente_id = c.cliente_id
JOIN productos p ON v.producto_id = p.producto_id
"""

pd.read_sql(query, conn)
```

### Clientes totales

```python
query = """
SELECT COUNT(*) AS total_clientes
FROM clientes;
"""

pd.read_sql(query, conn)
```

### Clientes por ciudad

```python
query = """
SELECT ciudad, COUNT(*) AS cantidad_clientes
FROM clientes
GROUP BY ciudad
ORDER BY cantidad_clientes DESC;
"""

pd.read_sql(query, conn)
```

### Ingresos totales

```python
query = """
SELECT SUM(total) AS ingresos_totales
FROM ventas;
"""

pd.read_sql(query, conn)
```

### Ingresos por ciudad

```python
query = """
SELECT c.ciudad, SUM(v.total) AS ingresos
FROM ventas v
JOIN clientes c ON v.cliente_id = c.cliente_id
GROUP BY c.ciudad
ORDER BY ingresos DESC;
"""

pd.read_sql(query, conn)
```

### Productos más vendidos (por cantidad)

```python
query = """
SELECT p.nombre_producto, SUM(v.cantidad) AS total_vendido
FROM ventas v
JOIN productos p ON v.producto_id = p.producto_id
GROUP BY p.nombre_producto
ORDER BY total_vendido DESC;
"""

pd.read_sql(query, conn)
```

### Productos que generan más ingresos

```python
query = """
SELECT p.nombre_producto, SUM(v.total) AS ingresos
FROM ventas v
JOIN productos p ON v.producto_id = p.producto_id
GROUP BY p.nombre_producto
ORDER BY ingresos DESC;
"""

pd.read_sql(query, conn)
```

### Ticket promedio por venta

```python
query = """
SELECT ROUND(AVG(total), 0) AS ticket_promedio
FROM ventas;
"""

pd.read_sql(query, conn)
```

### Clientes con mayor gasto

```python
query = """
SELECT c.nombre, SUM(v.total) AS gasto_total
FROM ventas v
JOIN clientes c ON v.cliente_id = c.cliente_id
GROUP BY c.nombre
ORDER BY gasto_total DESC;
"""

pd.read_sql(query, conn)
```

### Ventas por mes

```python
query = """
SELECT 
    strftime('%Y-%m', fecha_venta) AS mes,
    SUM(total) AS ingresos
FROM ventas
GROUP BY mes
ORDER BY mes;
"""

pd.read_sql(query, conn)
```

### Resumen ejecutivo

```python
query = """
SELECT
    COUNT(DISTINCT cliente_id) AS clientes_unicos,
    COUNT(*) AS total_ventas,
    SUM(total) AS ingresos_totales,
    ROUND(AVG(total), 0) AS ticket_promedio
FROM ventas;
"""

pd.read_sql(query, conn)
```
